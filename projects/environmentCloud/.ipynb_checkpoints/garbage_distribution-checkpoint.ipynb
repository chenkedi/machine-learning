```python
"""
读取单个片区并合并为一个整体片区文件。提取出其中的仓库输出为inputDepot.csv，其他的点位合并为inputeNode.csv
"""
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_excel(io="data/garbage.xlsx", header=0)
data.dropna(axis=0, how='all', inplace=True)
data.rename(columns=
            {'垃圾点名称':'id',
             '经度':'lng',
             '纬度':'lat',
             '装货重量（吨）':'weight',
             '送货序号':'order',
             '送货路线':'route'}, inplace=True)



station = data[data['id'].str.contains('中转站')]
# 单位处理时间 5 min
gabage = data[~data['id'].str.contains('中转站')]
gabage['serviceTime'] = 5

# processStation = data[data['facilityType'] == '处置端']

x = gabage['lng']
y = gabage['lat']
x2 = station['lng']
y2 = station['lat']
# x3 = processStation['lng']
# y3 = processStation['lat']

fig = plt.figure(figsize=(10, 10))
ax = plt.subplot()
ax.scatter(x, y, s=10, alpha=0.5) # 垃圾站
ax.scatter(x2, y2, s=100, c='green', alpha=0.6)  # 中转站
# ax.scatter(x3, y3, s=200, c='red') # 焚烧站
plt.show()
# gabage[['id','lng','lat','weight','serviceTime']].to_csv("data/inputNode.csv", index=False)
# station[['id','lng','lat']].to_csv("data/inputDepot.csv", index=False)
# 输出所有中转站和仓库的点位数据
data[['id', 'lng', 'lat']].to_excel("data/all_coordinate.xls", index=False)
```

```python
"""
唱凯片区转换为csv格式
"""
changkai = pd.read_excel("data/changkai_district.xlsx", dtype=object, names=['id', 'lng', 'lat', 'route', 'series', 'weight'])
changkai['serviceTime'] = 5
cur_res = changkai.iloc[1:]
changkai_district = changkai.iloc[1:][['id', 'lng', 'lat','weight','serviceTime']]
# changkai_district.to_csv("data/inputNodeSingle.csv", index=False)
```

```python
"""
分析唱凯片区我方结果与苏州环境云结果
"""
import decimal as dc
import numpy as np
dc.getcontext().rounding = dc.ROUND_HALF_UP

cur_res = cur_res[['id', 'route', 'weight', 'lng', 'lat']]
# print(cur_res)
matrix_id = pd.read_csv("data/location_changkai.csv", names=['id', 'lat', 'lng'], dtype={'id': int, 'lat': str, 'lng':str})
# , dtype={'id': int, 'lat': str, 'lng':str}
# matrix_id[['lat', 'lng']] = matrix_id[['lat', 'lng']].astype(str)
matrix_id = matrix_id.set_index(['lng', 'lat'])
# print(matrix_id)
# print(matrix_id['id'])


matrix = pd.read_csv("data/matrix_changkai.csv", names=['order', 'from_id', 'to_id', 'distance', 'time']).set_index(['from_id', 'to_id'])
# print(matrix['to_id'])
depot_id = matrix_id.loc['116.313324330542', '28.1209760513841']['id']

statistic = {}
for key, value in cur_res.groupby('route'):
    route_name = "路线%s" % key
    distance = 0.;
    weight = 0.;
    time = 0.;
    last_id = depot_id
    for i in range(len(value) - 1):
        cur_lng = str(value.iloc[i]['lng'])
        cur_lat = str(value.iloc[i]['lat'])
        
        cur_id = matrix_id.loc[cur_lng, cur_lat]['id']
       
        res = (0,0)
        if(last_id != cur_id):     
#             if(last_id in fromIdList):
#                 res = matrix.loc[last_id, cur_id][['distance', 'time']]
#             else:
#                 res = matrix.loc[cur_id, last_id][['distance', 'time']]
            try:
#                 print(type(cur_id), type(last_id))
                
#                 print("进入正常处理分支","cur_id: ", cur_id, "last_id: ", last_id)
                res = matrix.loc[last_id, cur_id][['distance', 'time']]
            except KeyError:
#                 print(type(cur_id), type(last_id))
#                 print("进入异常处理分支","cur_id: ", cur_id, "last_id: ", last_id)
#                 cur_id = int(21863)
#                 last_id = int(22125)
#                 print(matrix.loc[cur_id, last_id][['distance', 'time']])
                res = matrix.loc[cur_id, last_id][['distance', 'time']]
#         print(res)
        distance += res[0]
        time += res[1]
        time += 300 # 秒
        weight += value.iloc[i]['weight']
        last_id = cur_id
    
    #回仓
    res = (0, 0)
    if(last_id != depot_id):
#         if(last_id in fromIdList):
#             res = matrix.loc[last_id, depot_id][['distance', 'time']]
#         else:
#             res = matrix.loc[depot_id, last_id][['distance', 'time']]
        try:
            res = matrix.loc[last_id, depot_id][['distance', 'time']]
        except KeyError:
            res = matrix.loc[depot_id, last_id][['distance', 'time']]
    distance += res[0]
    time += res[1]
    statistic[route_name] = {'行驶距离':distance / 1000, '行驶时间':dc.Decimal(time / 60, dc.getcontext()).__round__(0), '垃圾重量':weight} 
    
final_res = pd.DataFrame(statistic)
final_res['总计'] = final_res.sum(axis = 1)
final_res
```

```python
total_time_improve = (final_res.loc['行驶时间']['总计'] - 2477) / 3157
max_time_improve = (468 - 389) / 468
print(max_time_improve)
total_distance_improve = (final_res.loc['行驶距离']['总计'] - 520.628) / 790.696

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

figure = plt.figure(figsize=(6, 6), dpi = 100)
opacity = 0.4

x = range(3)
y = [total_time_improve, max_time_improve, total_distance_improve]

ax = plt.subplot()
ax.bar(x, y, 0.25, color=["#87CEFA", "red", "green"], alpha=opacity)
plt.xticks(x, ['总行驶时间下降幅度', '最长行驶时间下降幅度', '总行驶距离下降幅度'], rotation=30)

for (a, b) in zip(x, y):
    plt.text(a, b+0.001, '%.2f%%' % (b*100), ha='center', va='bottom', fontsize=7)
plt.title("百度规划方案提升幅度统计")
```

```python
"""
将全部点位文件切分为单个分区
"""
import string

district_index_list = []
district = [0, 0]
for i in range(len(data) - 2):

    if(data.iloc[i]['id'].find('中转站') != -1):
        district[0] = i
        print()
    if(data.iloc[i + 1]['id'].find('中转站') != -1):
        district[1] = i
        district_index_list.append(district.copy())
district[1] = len(data) - 1
district_index_list.append(district)

district_list = []
for a, b in district_index_list:
    district_list.append(data.iloc[a:b].copy().reset_index(drop=True))
    
for d in district_list:
    pd.DataFrame(d.iloc[0][['id', 'lng', 'lat']]).T.to_csv("data/districts/inputDepot_%s.csv" % d.iloc[0]['id'], index=False)
    d.iloc[1:].to_csv("data/districts/inputNode_%s.csv" % d.iloc[0]['id'], index=False)
```
