```python
import data_visualize as dv
import load_data as ld
import matplotlib.pyplot as plt
%matplotlib inline

raw_data = ld.load_binary_data()
raw_data = raw_data[raw_data['系统效率'] < 6]
raw_data = raw_data[raw_data['系统效率'] > 0]
dv.visualize_all_data(raw_data)
```

```python
import nn_model as nn
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from keras.utils import plot_model

data_divid = 34000
epoch = 50
feature_domain = ['冷冻总管供水温度', '冷冻总管回水温度', '冷冻总管流量', '室内监测点温度', '室内监测点湿度','室外监测点温度', '室外监测点湿度', '室外监测点光照度', '冷冻泵总状态', '舒适度', 'control_level']
objective_domain = '系统效率'

X_train, scaler = nn.feature_normalize(raw_data[:data_divid][feature_domain].values)
Y_train = raw_data[:data_divid][objective_domain].values

X_test = scaler.transform(raw_data[data_divid:][feature_domain].values)
Y_test = raw_data[data_divid:][objective_domain].values

estimator = nn.model()
estimator.fit(X_train, Y_train, batch_size=32, epochs=epoch, verbose=1)
# estimator.load_weights('result/80%train-with-mean-normalize-50-epochs')
estimator.save_weights('result/80%train-with-mean-normalize-50-epochs-3')
            
# plot_model(estimator, to_file='model.png')
# SVG(model_to_dot(estimator).create(prog='dot', format='svg'))
print(estimator.evaluate(X_test, Y_test))
```

```python
Y_test_pred = estimator.predict(X_test)
X_test_timestamp = raw_data[data_divid:]['DateTime'].apply(lambda x: nn.datetime_timestamp(x)).values
# X_test_timestamp = raw_data.ix[data_divid:].index.values
plt.scatter(X_test_timestamp, Y_test)
plt.plot(X_test_timestamp, Y_test_pred, color='red')

plt.show()
```
